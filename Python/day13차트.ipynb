```python
import wikipedia
from wordcloud import WordCloud, STOPWORDS

wiki = wikipedia.page('Artificial intelligence')
text = wiki.content

wordcloud = WordCloud(width =2000, height = 1500).generate(text)
plt.figure(figsize=(40,30))
plt.imshow(wordcloud)
```

```python
wiki = wikipedia.page('South Korea')
wiki1 = wikipedia.page('Japan')
text = wiki.content+wiki1.content

s_word = STOPWORDS.union({'one', 'using', 'first', 'two', 'make', 'use', 'south', 'north', 'korean', 'world', 'country' })
wordcloud = WordCloud(width =2000, height = 1500,stopwords= s_word).generate(text)
plt.figure(figsize=(40,30))
plt.imshow(wordcloud)
```
