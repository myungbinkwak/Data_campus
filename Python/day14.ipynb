```python
max_temp = 0    #최고 기온값 저장할 변수
max_data = ''   #최고 기온이 가장 높았던 날짜를 저장
f = open('seoul.csv') 
data = csv.reader(f)
header = next(data)
for row in data:
    if row[-1] =='':
        row[-1] =-999   #-999를 넣어 빈 문자열이 있던 자리라고 표시
    row[-1] = float(row[-1])
    if max_temp < row[-1]:
        max_data = row[0]
        max_temp = row[-1]    
f.close()
print('기상 관측 이래 서울의  최고 기온이 가장 높았던 날은',max_data+'로, ',max_temp, '도 였습니다')
```

신도림에 있는 모든 연령을 리스트에 저장후 시각화

```python
f = open('age.csv', encoding='cp949')
data =csv.reader(f) 
result = []

for row in data:
    if'신도림' in row[0]:
        for i in row[3:]:
            result.append(int(i))
print(result)

plt.style.use('ggplot')
plt.plot(result)
```

아름동 지역의 인구구조

```python
a = input('인구 구조가 알고싶은 지역의 이름을 입력해주세요:')

f = open('age.csv', encoding='cp949')
data =csv.reader(f) 
result = []

for row in data:
    if a in row[0]:
        for i in row[3:]:
            result.append(int(i))
print(result)

plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.title(a+'지역의 인구구조')
plt.plot(result)
```

신도림에 사는 남자와 여자 0~100세

```python
a = open('gender.csv')
data = csv.reader(a)
m=[]
f=[]
for row in data:
    if '신도림' in row[0]:
        for i in range(0,101):
            m.append(int(row[i+3])
            f.append(int(row[-(i+1)]))

f.reverse()
print("m:",m)
print("f:",f)
```

```python
c = open('gender.csv')
data = csv.reader(c)
m=[]
f=[]
for row in data:
    if '신도림' in row[0]:
        for i in row[3,104]:
            m.append(-int(i))
        for i in row[106:]:
            f.append(int(i))
        break

plt.style.use('ggplot')
plt.figure(figsize = (10,5), dpi=300)
plt.barh(range(101), m)
plt.barh(range(101), f)
plt.legend()
plt.show()
```
