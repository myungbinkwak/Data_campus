```python
data = ({
    'name':['A','B','C','D','E','F','G'],
    'horse power':[130,250,190,300,210,220,170],
    'weight':[1.9,2.6,2.2,2.9,2.4,2.3,2.2],
    'efficiency':[16.3,10.2,11.1,7.1,12.1,13.2,14.2]
})
df['com'] = 'P'

df = pd.DataFrame(data)
print(df)

df = df.set_index('name')

print(df[df['horse power'] == df['horse power'].max()])

data1 = ({
    'name':['A','B','C','D'],
    'horse power':[120,220,120,220],
    'weight':[1.9,2.1,1.5,2.9],
    'efficiency':[18.3,19.2,21.1,17.3]
})
df1['com'] = 'Q'

df1 = pd.DataFrame(data1)
df1 = df1.set_index('name')

result = pd.concat([df, df1])

print(result)


result['hp x mile'] = result['horse power'] * result['efficiency']
result = result.set_index('name')
print(result.groupby('com').mean()['hp x mile'])
```
