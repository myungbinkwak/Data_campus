```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

x = np.arange(-4.5, 5, 0.5)
y = 2*x**3
plt.plot(x,y)
plt.grid(True)
```

```python
x = np.arange(0, 5, 1)
y1 = x
y2 = x + 1
y3 = x + 2
y4 = x + 3

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.plot(x,y4)
plt.grid(True)
plt.legend(['data1','data2','data3','data4'])
```

```python

x = [x for x in range(-10, 10)]
y = [2*t for t in x]
plt.plot(x,y, marker = 'o')
plt.axis([-20,20,-20,20])
```

```python
x = [x for x in range(-20, 20)]
y = [2*x for x in x]
y1 =[x**2+5 for x in x]
y2 =[-x**2-5 for x in x]
plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.axis([-30,30,-30,30])
```

```python
x = [x for x in range(20)] # 0에서 20까지의 정수를 생성
y = [x**2 for x in x] # 0에서 20까지의 정수 x에 대해 x 제곱값을 생성
z = [x**3 for x in x] # 0에서 20까지의 정수 x에 대해 x 세제곱값을 생성

plt.plot(x,x)
plt.plot(x,y)
plt.plot(x,z)
plt.legend(['linear', 'quadratic', 'qubic'])
plt.title('My plot')
```

```python
years = [1965, 1975, 1985, 1995, 2005, 2015]
ko = [130, 650, 2450, 11600, 17790, 27250]
jp = [890, 5120, 11500, 42130, 40560, 38780]
ch = [100, 200, 290, 540, 1760, 7940]

x = np.arange(len(years))
plt.bar(x+0.0, ko, width=0.25)
plt.bar(x+0.3, jp, width=0.25)
plt.bar(x+0.6, ch, width=0.25)
plt.xticks(range(len(years)),years)
```

```python
mu1, sigma1 = 10,2
mu2, sigma2 = -6,3
standard_Gauss = np.random.randn(10000)
Gauss1 = mu1 + sigma1*np.random.randn(10000)
Gauss2 = mu2 + sigma2*np.random.randn(10000)
plt.figure(figsize=(10,6))
plt.hist(standard_Gauss, bins=50, alpha = 0.4)
plt.hist(Gauss1, bins=50, alpha = 0.4)
plt.hist(Gauss2, bins=50, alpha = 0.4)
```
